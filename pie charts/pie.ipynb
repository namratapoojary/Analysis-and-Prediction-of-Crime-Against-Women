```python
import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('col_total.csv')
Year_val = df["crime"]
Crime = df["2001"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.2, 0, 0, 0,0 ,0,0)  
plt.pie(Crime, labels=Year_val, explode= None, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140, center=(0, 0) )
plt.axis('equal')
plt.title(" Crimes against women in the year 2001\n"+"")
plt.savefig('crime2001.png')
plt.show()
```
