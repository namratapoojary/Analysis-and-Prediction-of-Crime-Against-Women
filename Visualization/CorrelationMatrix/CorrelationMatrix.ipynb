```python
import numpy as np
import pandas as pd
import seaborn as sns
dataset =pd.read_csv('2001.csv')
variables_correlation = dataset[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']]
a = variables_correlation.corr()

b = sns.heatmap(a, xticklabels=a.columns,yticklabels=a.columns)
b.axis('equal')
c = b.get_figure()
c.savefig("CorrelationMatrix.png")
a
```
